```python
# multi-collinearity and low correlation both present in the data
# we will be using elastic net regression which combines both lasso and ridge regression
```

```python
# splitting the data kfoldcrossvalidation to give chance ofr every part of the data for evaluation
# kfoldcroassvalidation is gold technique to evaluate the model
```

```python
x.shape,y.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
xx = ss.fit_transform(x)
```

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.metrics import mean_absolute_error,make_scorer
from sklearn.linear_model import ElasticNet

num_folds = 8
seed = 42
kfold = KFold(n_splits=num_folds, random_state=seed,shuffle=True)
model = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=seed)
error = make_scorer(mean_absolute_error)
re = cross_val_score(model,x,y,cv = kfold,scoring=error)
```

```python
re.mean()
```

```python
y = y.values
```

```python
type(y),type(xx)
```

```python
# we will check whether our model is overfit or not by evaluating the training data and validation data for each fold  

kf = kfold
X_scaled = xx
train_scores = []
validation_scores = []

for train_index, val_index in kf.split(X_scaled):
    X_train, X_val = X_scaled[train_index], X_scaled[val_index]
    y_train, y_val = y[train_index], y[val_index]
    
    model.fit(X_train, y_train)
    
    # Calculate the training and validation MSE
    train_mse = mean_absolute_error(y_train, model.predict(X_train))
    val_mse = mean_absolute_error(y_val, model.predict(X_val))
    
    train_scores.append(train_mse)
    validation_scores.append(val_mse)

```

```python
import numpy as np
av1 = np.mean(train_scores)
av2 = np.mean(validation_scores)
print(av1,av2)
```

```python
print(train_scores)
print(validation_scores)
```

```python
# as we see the avg training scores are not significantly lower than avg validation scores, so we can say the the model is not a overfit nor underfit
```
